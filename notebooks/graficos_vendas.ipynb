```python
import matplotlib.pyplot as plt
import numpy as np

# Gerando dados aleatórios para exemplo
np.random.seed(42)
x = np.random.rand(50) * 100
y = 2.5 * x + np.random.randn(50) * 10

# Criando o gráfico de dispersão
plt.scatter(x, y, color='blue', alpha=0.7)

# Configuração do gráfico
plt.xlabel("Investimento em Marketing (milhares de R$)")
plt.ylabel("Receita Gerada (milhares de R$)")
plt.title("Correlação entre Investimento e Receita")

# Exibição do gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt

# Dados
categorias = ["Produto A", "Produto B", "Produto C", "Produto D"]
valores = [340, 410, 390, 450]

# Criando o gráfico de barras
plt.bar(categorias, valores, color=['blue', 'red', 'green', 'purple'])

# Configuração do gráfico
plt.xlabel("Produtos")
plt.ylabel("Quantidade Vendida")
plt.title("Vendas por Produto")

# Exibição do gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt

# Dados
empresas = ["Empresa A", "Empresa B", "Empresa C", "Empresa D"]
participacao = [35, 25, 20, 20]
cores = ["blue", "red", "green", "purple"]

# Criando o gráfico de pizza
plt.pie(participacao, labels=empresas, colors=cores, autopct="%1.1f%%", startangle=90)

# Configuração do gráfico
plt.title("Participação de Mercado")

# Exibição do gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt

# Dados
anos = [2018, 2019, 2020, 2021, 2022]
vendas = [150, 180, 220, 210, 250]

# Criando o gráfico com estilo
plt.style.use('ggplot')
plt.plot(anos, vendas, marker='o', linestyle='-', color='b', label="Vendas")

# Configuração do gráfico
plt.xlabel("Ano")
plt.ylabel("Vendas")
plt.title("Evolução das Vendas com Destaque")
plt.legend()

# Exibição do gráfico
plt.show()
```
